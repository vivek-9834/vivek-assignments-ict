```python
x = data[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y = data['Outcome']
```

model creation

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
#create model by logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred = log.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print('accuracy_score = ',accuracy_score(y_test,y_pred))
print('Precision = ',precision_score(y_test,y_pred))
print('Recall_score = ',recall_score(y_test,y_pred))
print('f1 score = ',f1_score(y_test,y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
```

```python
model1 = LogisticRegression()
model1.fit(x_train,y_train)
```

```python
y_pred_sc = model1.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred_sc)
```

```python
accuracy_score(y_test,y_pred_sc)
```

```python
#using KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metrik_k = []
neighbors = np.arange(3,15)

for k in neighbors:
  classifier = KNeighborsClassifier(n_neighbors=k)
  classifier.fit(x_train,y_train)
  y_pred = classifier.predict(x_test)
  acc = accuracy_score(y_test,y_pred)
  metrik_k.append(acc)
```

```python
metrik_k
```

```python
plt.plot(neighbors,metrik_k,'o-')
plt.xlabel('K value')
plt.ylabel('Accuracy')
plt.grid()
plt.show()
```

```python
classifier = KNeighborsClassifier(n_neighbors=8)
classifier.fit(x_train,y_train)
```

```python
y_pred_knn = classifier.predict(x_test)
accuracy_score(y_test,y_pred_knn)
```

```python
metrik_k = []
neighbors = np.arange(3,15)

for k in neighbors:
  classifier = KNeighborsClassifier(n_neighbors=k)
  classifier.fit(x_train_sc,y_train)
  y_pred = classifier.predict(x_test_sc)
  acc = accuracy_score(y_test,y_pred)
  metrik_k.append(acc)
```

```python
plt.plot(neighbors,metrik_k,'o-')
plt.xlabel('K value')
plt.ylabel('Accuracy')
plt.grid()
plt.show()
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train_sc,y_train)
```

```python
y_pred_knn_sc = classifier.predict(x_test_sc)
accuracy_score(y_test,y_pred_knn_sc)
```
