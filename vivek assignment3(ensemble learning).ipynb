```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('/content/train_loan (2).csv')
#preprosessing steps
data.head(10)
```

```python
data['Dependents']=data['Dependents'].replace('3+','3')
data['Dependents'].unique()
```

missing value handling

```python
data['Gender']=data['Gender'].fillna(data['Gender'].mode()[0])
data['Married']=data['Married'].fillna(data['Married'].mode()[0])
data['Dependents']=data['Dependents'].fillna(data['Dependents'].mode()[0])
data['Self_Employed']=data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
data['LoanAmount']=data['LoanAmount'].fillna(data['LoanAmount'].mean())
data['Loan_Amount_Term']=data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mean())
data['Credit_History']=data['Credit_History'].fillna(data['Credit_History'].mean())
```

```python
data.isna().sum()
```

encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
```

```python
data['Married']=le.fit_transform(data['Married'])
```

```python
data['Education']=le.fit_transform(data['Education'])
```

```python
data['Self_Employed']=le.fit_transform(data['Self_Employed'])
```

```python
data['Dependents']=le.fit_transform(data['Dependents'])
```

```python
data['Property_Area']=le.fit_transform(data['Property_Area'])



```

```python
data['Loan_ID']=le.fit_transform(data['Loan_ID'])
```

```python
data.info()

```

```python
x=data.drop('Loan_Status',axis=1)
y=data['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
```

Using Randomforest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
accuracy_score(y_test,y_pred_rf)
```

```python
#bagging classifier
from sklearn.ensemble import BaggingClassifier
bg_clf=BaggingClassifier(n_estimators=100,random_state=42)
bg_clf.fit(x_train,y_train)
y_pred_bg=bg_clf.predict(x_test)
accuracy_score(y_test,y_pred_bg)
```

```python
#bagging classifier with eleminator SVC
```

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bg_clf=BaggingClassifier(estimator=svc,n_estimators=100,random_state=42)
bg_clf.fit(x_train,y_train)
y_pred_bg=bg_clf.predict(x_test)
accuracy_score(y_test,y_pred_bg)
```

```python
#adaptive boosting
from sklearn.ensemble import AdaBoostClassifier
ada_boost=AdaBoostClassifier(n_estimators=100)
ada_boost.fit(x_train,y_train)
y_pred_ada=ada_boost.predict(x_test)
accuracy_score(y_test,y_pred_ada)
```

```python
#Gradiant boosting
from sklearn.ensemble import GradientBoostingClassifier
gb_clf=GradientBoostingClassifier(n_estimators=100,random_state=42,max_depth=5)
gb_clf.fit(x_train,y_train)
y_pred_gb=gb_clf.predict(x_test)
accuracy_score(y_test,y_pred_gb)
```

```python

#XG Boosting
import xgboost as xgb
xgb_model=xgb.XGBClassifier(n_estimators=100,random_state=42,learning_rate=.5)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.fit_transform(y_test)
xgb_model.fit(x_train,y_train)
y_pred_xgb=xgb_model.predict(x_test)
accuracy_score(y_test,y_pred_xgb)
```
